```python
import numpy as np
import matplotlib.pyplot as plt
```

# Minimasi Loss Function Satu Feature (Linear Loss Function Minimization)

```python
np.random.seed(0)
t=np.random.randn(100)
x=np.linspace(0,100,100)
```

```python
x_bar = np.mean(x)
t_bar = np.mean(t)
xt_bar = np.mean(x*t)
x2_bar = np.mean(x**2)
```

```python
w1 = (xt_bar - x_bar*t_bar)/(x2_bar - x_bar**2)
w0 = t_bar - w1*x_bar
```

```python
plt.clf()
plt.plot(x,t)
```

```python
y=w1*x+w0 #y adalah t hasil model linear
print(w0)
print(w1)
```

```python
plt.clf()
plt.plot(x,y)
plt.plot(x,t)
```

# Minimasi Loss Function Satu Feature dengan Vector

```python
N = t.shape[0]
X = np.zeros((t.shape[0],2))
```

```python
X[:,0] = np.ones(N)
```

```python
X[:,1] = x
```

```python
XtX = np.matmul(np.transpose(X),X)
XtXinv = np.linalg.inv(XtX)
```

```python
w = np.matmul(np.matmul(XtXinv,np.transpose(X)),t)
```

```python
w
```

```python
t_new = np.zeros(t.shape[0])
for i in range (t.shape[0]):
    t_new[i] = np.matmul(w,np.transpose(X[i,:]))
```

```python
plt.clf()
plt.plot(x,t_new)
plt.plot(x,t)
```

# Minimasi Loss Function Banyak Feature (dipilih orde 3) dengan Vector

```python
N = t.shape[0]
X = np.zeros((t.shape[0],4))
X[:,0] = np.ones(N)
X[:,1] = x
X[:,2] = x*x
X[:,3] = x*x*x
XtX = np.matmul(np.transpose(X),X)
XtXinv = np.linalg.inv(XtX)
w = np.matmul(np.matmul(XtXinv,np.transpose(X)),t)
```

```python
w
```

```python
t_new = np.zeros(t.shape[0])
for i in range (t.shape[0]):
    t_new[i] = np.matmul(w,np.transpose(X[i,:]))
```

```python
plt.clf()
plt.plot(x,t_new)
plt.plot(x,t)
```

# Minimasi Loss Function Banyak Feature (dipilih orde 3) dengan Regularisasi dan Vector

```python
lam=100
N = t.shape[0]
X = np.zeros((t.shape[0],4))
X[:,0] = np.ones(N)
X[:,1] = x
X[:,2] = x*x
X[:,3] = x*x*x
XtX = np.matmul(np.transpose(X),X)
lam=lam*np.identity(XtX.shape[0])
NlamI=N*lam
w = np.matmul(np.matmul(np.linalg.inv(XtX+NlamI),np.transpose(X)),t)
```

```python
w
```

```python
t_new = np.zeros(t.shape[0])
for i in range (t.shape[0]):
    t_new[i] = np.matmul(w,np.transpose(X[i,:]))
```

```python
plt.clf()
plt.plot(x,t_new)
plt.plot(x,t)
```
