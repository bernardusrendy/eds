# REGRESI LINIER 

```python
import numpy as np 
from matplotlib import pyplot as plt
x = np.array([[0.0],[1.0],[2.0]])
y = np.array([1.0,2.0,2.9]) 
plt.scatter(x,y)
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression().fit(x, y) 
print("m =",lm.coef_, "; dan b =", lm.intercept_)
#lm.coef_ # m array([ 0.95]) 
#lm.intercept_  # b 1.0166666666666671 
```

```python
# y_pred = mx + b
y_pred = lm.predict(x) 
plt.scatter(x,y)
plt.plot(x,y_pred)
```

```python
from sklearn.metrics import r2_score
print('R^2 =', r2_score(y_pred,y))
```

# REGRESI NON-LINIER (POLINOM)

```python
import operator

import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import PolynomialFeatures

np.random.seed(0)
x = 2 - 3 * np.random.normal(0, 1, 20)
y = x - 2 * (x ** 2) + 0.5 * (x ** 3) + np.random.normal(-3, 3, 20)

# transforming the data to include another axis
x = x[:, np.newaxis]
y = y[:, np.newaxis]

degr = int(input("degree polinom (int)= "))
polynomial_features= PolynomialFeatures(degree=degr)
x_poly = polynomial_features.fit_transform(x)

model = LinearRegression()
model.fit(x_poly, y)
y_poly_pred = model.predict(x_poly)

rmse = np.sqrt(mean_squared_error(y,y_poly_pred))
r2 = r2_score(y,y_poly_pred)
print("RMSE =", rmse)
print("R^2 =", r2)

plt.scatter(x, y, s=10)
# sort the values of x before line plot
sort_axis = operator.itemgetter(0)
sorted_zip = sorted(zip(x,y_poly_pred), key=sort_axis)
x, y_poly_pred = zip(*sorted_zip)
plt.plot(x, y_poly_pred, color='m')
plt.show()
```
